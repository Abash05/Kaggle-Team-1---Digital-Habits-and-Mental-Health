```python
sns.scatterplot(
    data=df,
    x='anxiety_score',
    y='productivity_score',
    alpha=0.5
)
sns.regplot(
    data=df,
    x='anxiety_score',
    y='productivity_score',
    scatter=False,
    color='red',
    line_kws={"lw": 2}
)
plt.title("Relationship between Anxiety Score and Productivity Score", fontsize=14, fontweight="bold")
plt.xlabel("Anxiety Score")
plt.ylabel("Productivity Score")
plt.grid(alpha=0.3)
plt.show()
```
