# Focus Score vs other collumns
focus had the strongest correlations with notifications per day, social meadia minutes, and digital dependence scores

```python
numeric_collumns = ['id','age','device_hours_per_day','phone_unlocks','notifications_per_day','social_media_mins',
                    'study_mins','physical_activity_days','sleep_hours','sleep_quality','anxiety_score','depression_score',
                    'stress_level','happiness_score','focus_score','high_risk_flag','productivity_score','digital_dependence_score']
                    
focus_corr = df_dl[numeric_collumns].corr()['focus_score'].round(2).to_frame().style.background_gradient(cmap='coolwarm')
focus_corr

#Highest correlations
high_corr_collumns = ['focus_score','notifications_per_day','social_media_mins','digital_dependence_score','stress_level']
df_dl_highcorr = df_dl[high_corr_collumns]
df_dl_highcorr.corr()['focus_score'].round(2).to_frame().style.background_gradient(cmap='coolwarm')
```

```python
import seaborn as sb

# Daily Notifications vs Focus Score
plt.figure(figsize=(5, 5))
sb.regplot(
    x='notifications_per_day', 
    y='focus_score',
    data=df_dl, 
    ci=95,     
    scatter_kws={'alpha': 0.7, 'color': 'blue'}, 
    line_kws={'color': 'red'}              
)

plt.title('Daily Notifications vs Focus Score')
plt.xlabel('Notifications')
plt.ylabel('Focus Score', fontsize=12)
plt.show()

# Digital Dependence Score vs Focus Score
plt.figure(figsize=(5,5))
sb.regplot(
    x='digital_dependence_score', 
    y='focus_score',
    data=df_dl, 
    ci=95,       
    scatter_kws={'alpha': 0.7, 'color': 'blue'}, 
    line_kws={'color': 'red'}                   
)

plt.title('Digital Dependence Score vs Focus Score')
plt.xlabel('Digital Dependence Score')
plt.ylabel('Focus Score', fontsize=12)
plt.show()

# Social Media Minutes vs Focus Score
plt.figure(figsize=(5,5))
sb.regplot(
    x='social_media_mins', 
    y='focus_score',
    data=df_dl, 
    ci=95,       
    scatter_kws={'alpha': 0.7, 'color': 'blue'}, 
    line_kws={'color': 'red'}                   
)

plt.title('Social Media Minutes vs Focus Score')
plt.xlabel('Social Media Minutes')
plt.ylabel('Focus Score', fontsize=12)
plt.show()
```

```python
well_being_cols = ['anxiety_score', 'depression_score', 'stress_level', 'happiness_score',
                   'focus_score', 'productivity_score', 'digital_dependence_score', 'high_risk_flag']

subset_corr = df_dl[well_being_cols].corr()

plt.figure(figsize=(8, 7))
sb.heatmap(subset_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Well-Being Factor Correlations')
plt.show()
```

Digital dependence has high correlations with poor wellbeing seen through anxiety, depression, and stress correlation and negative correlations to negative things like hapiness and focus. All this is clearly seen in the second to last row of the matrix above.
The high risk flag also seems to have some correlation with digitial dependence.
